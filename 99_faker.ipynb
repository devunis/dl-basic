```python
import pandas as pd
import random

fake = Faker()

def gen_customer(n = 1000):
    arr = []
    for _ in range(n):
        data = {
            "AGE": fake.random_int(min=18, max=90),
            "SUB_MONTHS": fake.random_int(min=1, max=120),
            "MONTHLY_PAYME?NT": round(random.uniform(10, 100), 2),
            "USAGE_FEQ": fake.random_int(min=1, max=30),
            "STFT": fake.random_int(min=1, max=5),
            "CHURN": fake.boolean(chance_of_getting_true=30)
        }
        arr.append(data)
    return pd.DataFrame(arr)
```

```python
df = gen_customer(200000)
df
```

```python
df.to_csv("customer.csv", index=False)
```
